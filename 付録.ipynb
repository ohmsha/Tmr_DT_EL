### コードA.1　乱数

```python
import numpy as np
import matplotlib.pyplot as plt
# シードを指定すると毎回同じ乱数を生成する
np.random.seed(42)
# 0以上1未満の乱数を4個生成し配列を生成
J = np.random.rand(4)
print(J)
```

```python
# 0以上1未満の乱数を1000000個生成しヒストグラムを表示
K = np.random.rand(1000000)# 300分割してヒストグラムを生成
plt.hist(K, bins=300)
plt.show()
```

```python
# 標準正規分布の乱数を1000000個生成しヒストグラムを表示
L = np.random.randn(1000000)
# 300分割してヒストグラムを生成
plt.hist(L, bins=300)
plt.show()
```

```python
# 0以上1未満の乱数で2×300の配列を生成
M = np.random.rand(2, 300)# 0以上1未満の乱数を座標として散布図作成
X = M[0]
Y = M[1]
plt.scatter(X, Y)
plt.show()
```

```python
# 標準正規分布乱数を座標として散布図作成
N = np.random.randn(2, 300)
X = N[0]
Y = N[1]
plt.scatter(X, Y)
plt.show()
```

### コードA.4　DataFrameからの描画

```python
import pandas as pd
import matplotlib.pyplot as plt
# 描画領域を作成
plt.figure()
# csvファイルをDataFrameに読み込ませる
df_iris = pd.read_csv(
'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
)
# x,y座標の列を指定しsetosa個体を四角形でプロット
ax1 = df_iris.loc[df_iris['species'] == 'setosa'].plot(kind='scatter',x='petal_length',y='sepal_length',marker='s')
# x,y座標の列を指定しversicolor個体を三角形でプロット
df_iris.loc[df_iris['species'] == 'versicolor'].plot(kind='scatter',x='petal_length',y='sepal_length',marker='^',
ax = ax1)
```

### コードA.5　グラフ作成の基本形

```python
import numpy as np
import matplotlib.pyplot as plt
# グラフ化する関数の定義
x = np.arange(0, 2.05, 0.05)
y = 1 - (x - 1)**2
# グラフをプロット
plt.plot(x, y)
# グラフを表示
plt.show()
```

### コードA.6　pyplotを活用してグラフ作成

```python
import numpy as np
import matplotlib.pyplot as plt
# 図示する三つの関数を定義
x = np.arange(0, 2.05, 0.05)
y1 = 1 - (x - 1)**2
y2 = 0.25 * x**2
y3 = 0.5 * np.sin(2 * np.pi * x / 2) + 0.5
# 描画領域を作成
plt.figure(figsize=(18,16))
# グラフ上に書く文字のフォントサイズ設定
plt.rcParams["font.size"] = 24
# グラフの属性を細かく定義
def graphconfig():
    # 凡例を表示する位置を左上にする
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    # 補助目盛を表示する
    plt.minorticks_on()
    # x座標の目盛りを定義
    plt.xticks(np.arange(0, 2.05, 0.5))
    plt.yticks(np.arange(0, 1.05, 0.5))
    # 座標軸の名前を設定
    plt.xlabel(r'$x$')
    plt.ylabel(r'$y$')
    # 主目盛のグリッド線の形状，色を設定
    plt.grid(which='major', linestyle='-', color='k')
    # 補助目盛のグリッド線の形状，色を設定
    plt.grid(which='minor', linestyle='dotted', color='k')
# 2行1列の1行目に描画領域設定
plt.subplot(211)
# 関数y1,y2のグラフをプロット
plt.plot(x, y1, label= r'$y_1=1-(x-1)^2$', color='g')
plt.plot(x, y2, label= r'$y_2=\frac{1}{4}x^2$', color='r')
# グラフ属性を設定する関数を呼び出す
graphconfig()
# 2行1列の2行目に描画領域設定
plt.subplot(212)
# 関数y2,y3のグラフをプロット
plt.plot(x, y2, label= r'$y_2=\frac{1}{4}x^2$', color='r')
plt.plot(x, y3, label= r'$y_3=\frac{1}{2}sin\frac{2\pi x}{2}+\frac{1}{2}$', color='b')
# グラフ属性を設定する関数を呼び出す
graphconfig()
# レイアウトの余白を少なめに設定
plt.tight_layout()
# グラフを表示
plt.show()
```
