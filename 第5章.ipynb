# 第5章　ランダムフォレスト

## 5.4　Pythonでランダムフォレスト

### コード5.1

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# アヤメデータの読み込みと特徴量選択（2つの特徴量のみを使用）
iris = load_iris()

# インデックス: 0=sepal length, 1=sepal width, 2=petal length, 3=petal width
feature_idx_1 = 2  # X軸: petal length
feature_idx_2 = 3  # Y軸: petal width

# 指定したインデックスの列を抽出
X = iris.data[:, [feature_idx_1, feature_idx_2]]
y = iris.target

# 指定したインデックスに対応する特徴量名を取得
x_label = iris.feature_names[feature_idx_1]
y_label = iris.feature_names[feature_idx_2]

# データ分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ランダムフォレストモデルのトレーニング
model = RandomForestClassifier(
    n_estimators=300,
    max_features=1,
    max_depth=None,
    random_state=42
)

model.fit(X_train, y_train)

# テストデータで予測
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# メッシュグリッドの作成（分類境界を描画するため）
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                    np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# クラスごとに異なるマーカーでプロット
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Greys)  # 分類境界の描画（薄いグレーの背景）

markers = ['o', 's', '^']  # 各クラスのマーカー形状
for i, marker in enumerate(markers):
    plt.scatter(X[y == i, 0], X[y == i, 1],
                label=iris.target_names[i], marker=marker, edgecolor='black')

# グラフの設定
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.legend(loc='best')
plt.title(f'Random Forest Classification of Iris Dataset ({x_label} vs {y_label})')
plt.grid(True)

# SVG形式で保存
output_path = f"random_forest_classification_iris_{x_label}_{y_label}_with_boundary.svg"
plt.savefig(output_path, format='svg')
print(f'Plot saved as SVG at {output_path}')

# 結果を表示
plt.show()
```
